```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
#LOF
ROC1_train = [1.0, 1.0, 0.9933, 0.548]
ROC1_test = [1.0, 1.0, 0.95, 0.4744]
ACC1_train = [0.91, 0.95, 0.97, 0.78]
ACC1_test = [0.92, 0.99, 0.97, 0.76]

#iForest
ROC3_train = [1.0, 1.0, 1.0, 1.0]
ROC3_test = [1.0, 1.0, 1.0, 1.0]
ACC3_train = [0.91, 0.95, 1.0, 0.9]
ACC3_test = [0.94, 0.96, 0.99, 0.93]
#EIF(Threshold=0.48,0.48,0.48,0.48)
ROC4_train = [0.9798, 0.9974, 1.0, 0.95]
ROC4_test = [0.9747, 1.0, 1.0, 1.0]
ACC4_train = [0.96, 0.995, 1.0, 0.98]
ACC4_test = [0.95, 1.0, 1.0, 1.0]
#DIF(threshold=0.30,0.30,0.30,0.30)
ROC5_train = [0.9848, 1.0, 1.0, 0.9]
ROC5_test = [0.9848, 1.0, 0.95, 0.975]
ACC5_train = [0.975, 0.99, 1.0, 1.0]
ACC5_test = [0.92, 0.98, 1.0, 1.0]
```

```python
labels = ['0.01', '0.05', '0.1', '0.2']

plt.plot(x, ROC1_train,label='LOF')
plt.plot(x, ROC3_train,label='iForest')
plt.plot(x, ROC4_train,label='EIF')
plt.plot(x, ROC5_train,label='DIF')
plt.xticks(x, labels)
plt.margins(0.2)
plt.title("ROC with Training Data by Outlier")
plt.legend(loc='lower left')
plt.show()


plt.plot(x, ROC1_test,label='LOF')
plt.plot(x, ROC3_test,label='iForest')
plt.plot(x, ROC4_test,label='EIF')
plt.plot(x, ROC5_test,label='DIF')
plt.xticks(x, labels)
plt.margins(0.2)
plt.title("ROC with Testing Data by Outlier")
plt.legend(loc='lower left')
plt.show()


plt.plot(x, ACC1_train,label='LOF')
plt.plot(x, ACC3_train,label='iForest')
plt.plot(x, ACC4_train,label='EIF')
plt.plot(x, ACC5_train,label='DIF')
plt.xticks(x, labels)
plt.margins(0.2)
plt.title("Accuracy with Training Data by Outlier")
plt.legend(loc='lower left')
plt.show()


plt.plot(x, ACC1_test,label='LOF')
plt.plot(x, ACC3_test,label='iForest')
plt.plot(x, ACC4_test,label='EIF')
plt.plot(x, ACC5_test,label='DIF')
plt.xticks(x, labels)
plt.margins(0.2)
plt.title("Accuracy with Testing Data by Outlier")
plt.legend(loc='best')
plt.show()
```
